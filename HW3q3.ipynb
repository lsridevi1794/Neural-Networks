# Linear least Squares 

Implementation of the linear least squares method: The weights that minimize the given cost function is computed and a linear fit to the given points is plotted. 


<p style="page-break-after:always;"></p>

```python
import numpy as np
import random as rand
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,51).reshape(1,50)  #Setting x from 1 to 50
one=np.ones([1,50])
X=np.array([x,one]).reshape(2,50)
```

```python
y=x+np.random.uniform(-1,1,size=[1,50]) #Setting y to be x+arbitrary real number
P=np.matmul(np.transpose(X),np.linalg.inv(np.matmul(X,np.transpose(X)))) #Calculating pseudo inverse
w=np.matmul(y,P) #Obtaining the weights
```

```python
plt.scatter(x,y)
z=(w[0][0]*x)+w[0][1]
plt.plot(x,z,'r.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Least Squares Plot')
plt.show()
```
