```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt

xi=np.random.uniform(0,1,300)
o=np.ones(300)
x=np.column_stack((xi,o))
v=np.random.uniform(-0.1,0.1,300)
D=[]
for i in range(300):
    d=math.sin(20*xi[i])+3*xi[i]+v[i]
    D.append(d)
```

```python
plt.plot(xi,D,'b.')
plt.show
```

```python
#w0=np.random.uniform(-1,1,48).reshape(24,2)
w0=np.random.randn(48).reshape(24,2)
w1=np.random.randn(24)
#w1=np.random.uniform(-1,1,24)
b=np.random.randn()
z0=np.zeros(24)
z=np.zeros(24)
z1=0
y=np.zeros(len(xi))
mse=np.zeros(300)
epoch=0
eta=0.1
MSE=[]
Epoch=[]
```

```python
while(epoch<2 or abs(MSE[epoch-1]>=0.2)):

    for i in range(len(xi)):
        for j in range(24):
            for k in range(0,2):
                z0[j]=np.matmul(w0[j],x[i].reshape(2,1))
                z[j]=np.tanh(np.matmul(w0[j],x[i].reshape(2,1)))
                z1=np.matmul(z.reshape(1,24),w1)+b
                y[i]=z1
                g=-(D[i]-y[i])*w1.reshape(24,1)*(1-np.square(np.tanh(z0[j])))
                dg1=g*xi[i]
                db1=g
                dg2=(-(D[i]-y[i])*z).reshape(1,24)
                db2=-(D[i]-y[i])
                G=np.array([dg1,db1]).reshape(2,24)
                w0[j][k]=w0[j][k]-(eta*G[k][j])
                w1[j]=w1[j]-(eta*dg2[0][j])
                b=b-(eta*db2)                
                

    epoch=epoch+1
    Epoch.append(epoch)
    for i1 in range(len(xi)):
        for j1 in range(24):
            for k1 in range(0,2):
                    

                z[j1]=np.tanh(np.matmul(w0[j1],x[i1].reshape(2,1)))
                z1=np.matmul(z.reshape(1,24),w1)+b
                y[i1]=z1  
                mse[i1]=(D[i1]-y[i1])*(D[i1]-y[i1])
    
    MSE.append(sum(mse)/300)
    if MSE[epoch-1]>=MSE[epoch-2]:
        eta-=0.01*eta
        print('Learning_rate: ',eta)
    
    print(MSE[epoch-1])

```

```python
plt.plot(Epoch,MSE)
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.show
```
